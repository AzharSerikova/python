```python
with open('text.txt') as f:
    txt = f.read()
```

```python
len(txt) #len or sum of our text
```

```python
d = {} # creating dictionary and counting each char
for x in txt:
     d[x] = d.get(x, 0) + 1 
```

```python
d
```

```python
prob_dict = {} # caclulating probability of each char
print('Char','|', 'Count','|Probobility')
print('----------------------------------')
for a, b in d.items():
    prob_dict[a] =( b / sum(d.values()))*100
    print(a, '    ',b, '    ',prob_dict[a])
```
